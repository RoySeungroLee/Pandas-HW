```python
% matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "raw_data/city_data.csv"
file2 = "raw_data/ride_data.csv"
```

```python
df = pd.read_csv(file)
df.head(10)
```

```python
df2 = pd.read_csv(file2)
df2.head()
```

```python
merge_df = pd.merge(df2,df,on="city",how="left")
merge_df
```

```python
group_city = merge_df.groupby("city")
group_count = group_city.count()
total = group_count["ride_id"].sum()
```

```python
group_city2 = merge_df.groupby("city")
city_mean = group_city2["fare"].mean()
```

```python
group_city3 = merge_df.groupby(["city","driver_count","type","fare"])
merge_df2 = group_city3.count().reset_index()
merge_df2.head(50)
```

```python
group_by = merge_df.groupby("city")
x = group_by["date"].count()
y = group_by["fare"].mean()
z = group_by["driver_count"].count()
plt.scatter(x,y , s=z, c="red", alpha=0.4)
plt.title("Pyber Ride Sharing Data(2018)")
```

```python
group_type = merge_df.groupby(["type"])
group_type_count = group_type.count()
group_type_count["city"].reset_index()
group_type_count["percent"] = (group_type_count["city"]/group_type_count["city"].sum())*100
group_type_count.plot(kind="pie",x="type", y ="percent", label="percent")
plt.title("% of Total Rides by City Type")
```

```python
group_type = merge_df.groupby(["type"])
group_type_sum = group_type.sum()
group_type_sum["fare"].reset_index()
group_type_sum["percent"] = (group_type_sum["fare"]/group_type_sum["fare"].sum())*100
graph = group_type_sum.plot(kind="pie",x="type",y="percent")
plt.title("% of Total Fares by City Type")
#group_type_count = group_type.count()
#group_type_count["city"].reset_index()
#group_type_count.plot(kind="pie",x="type", y ="city")
```
